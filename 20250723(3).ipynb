```python
# ‚úÖ 1. ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from rdkit import Chem
from rdkit.Chem import Descriptors, AllChem, rdFMCS
from rdkit.DataStructs import TanimotoSimilarity
from rdkit.Chem import rdMolDescriptors

from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
from sklearn.decomposition import PCA

sns.set_style('whitegrid')
```

```python
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ Î∞è Mol Î≥ÄÌôò
# Colab ÌôòÍ≤ΩÏóê train.csvÍ∞Ä ÏûàÎã§Í≥† Í∞ÄÏ†ï
# Í≤ΩÎ°ú Ïòà: /content/data_1/train.csv
df = pd.read_csv("/content/data_1/train.csv")  # Í≤ΩÎ°ú Ï°∞Ï†ï ÌïÑÏöî Ïãú Î≥ÄÍ≤Ω
df["mol"] = df["Canonical_Smiles"].apply(lambda smi: Chem.MolFromSmiles(smi))
df = df[df["mol"].notnull()].reset_index(drop=True)

print(f"‚úÖ Ïú†Ìö®Ìïú Mol Ïàò: {len(df)}")
```

```python
# MCS Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞ Ìï®Ïàò
def compute_mcs_similarity(mols):
    n = len(mols)
    sim = np.eye(n)
    for i in range(n - 1):
        for j in range(i + 1, n):
            res = rdFMCS.FindMCS([mols[i], mols[j]])
            na = res.numAtoms
            a1, a2 = mols[i].GetNumAtoms(), mols[j].GetNumAtoms()
            sim_val = na / (a1 + a2 - na) if (a1 + a2 - na) > 0 else 0
            sim[i, j] = sim[j, i] = sim_val
    return sim
```

```python
# Ïú†ÏÇ¨ÎèÑ ÌûàÌä∏Îßµ & PCA ÏãúÍ∞ÅÌôî Ìï®Ïàò
def plot_heatmap(sim_matrix, ids, title, threshold=None, method='single'):
    dist = 1.0 - sim_matrix
    condensed = squareform(dist, checks=False)
    Z = linkage(condensed, method=method)

    sns.clustermap(
        pd.DataFrame(sim_matrix, index=ids, columns=ids),
        row_linkage=Z, col_linkage=Z,
        cmap='vlag', figsize=(10, 10)
    )
    plt.suptitle(title, y=1.02, fontsize=14)
    plt.show()

    if threshold is not None:
        plt.figure(figsize=(12, 4))
        dendrogram(Z, no_labels=True)
        plt.axhline(y=threshold, color='red', linestyle='--')
        plt.title(f"Dendrogram (threshold={threshold})", fontsize=12)
        plt.xlabel('Distance')
        plt.ylabel('Cluster')
        plt.show()

        clusters = fcluster(Z, t=threshold, criterion='distance')
        counts = pd.Series(clusters).value_counts()
        print("Clusters (size > 5):")
        print(counts[counts > 5])

def plot_pca(sim_matrix, ids, inhibition, cmap='viridis', title='PCA on Similarity'):
    pcs = PCA(n_components=2).fit_transform(sim_matrix)
    df_pca = pd.DataFrame(pcs, index=ids, columns=['PC1', 'PC2'])
    df_pca['Inhibition'] = inhibition.values

    plt.figure(figsize=(8, 6))
    sns.scatterplot(
        data=df_pca, x='PC1', y='PC2',
        hue='Inhibition', palette=cmap,
        edgecolor='none'
    )
    plt.title(title, fontsize=14)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.legend(title='Inhibition', bbox_to_anchor=(1.05, 1))
    plt.tight_layout()
    plt.show()
```

```python
sample_df = df.sample(50, random_state=42).copy()  # ÎÑàÎ¨¥ ÎßéÏúºÎ©¥ Í≥ÑÏÇ∞ Ïò§Îûò Í±∏Î¶º
mols = sample_df["mol"].tolist()
sim_matrix = compute_mcs_similarity(mols)

# ÏãúÍ∞ÅÌôî
plot_heatmap(sim_matrix, sample_df["ID"].astype(str).tolist(), "üìå MCS Ïú†ÏÇ¨ÎèÑ ÌûàÌä∏Îßµ", threshold=0.4)
plot_pca(sim_matrix, sample_df["ID"].astype(str).tolist(), sample_df["Inhibition"], title="üìå PCA on MCS Similarity")

# Îç¥ÎìúÎ°úÍ∑∏Îû®(Dendrogram)ÏùÑ Ïù¥Ïö©Ìïú Íµ¨Ï°∞Ï†Å Ïú†ÏÇ¨ÎèÑ Í∏∞Î∞ò Í≥ÑÏ∏°Ï†ï ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Î∂ÑÏÑù
# XÏ∂ï Í∞ÅÍ∞ÅÏùò Î∂ÑÏûê ÏÉòÌîåÎì§ (ÎòêÎäî Ïú†ÏÇ¨Ìïú ÏÉòÌîåÎÅºÎ¶¨ Î¨∂Ïù∏ Í∑∏Î£π)
# YÏ∂ï (Distance): ÏÉòÌîå ÎòêÎäî ÌÅ¥Îü¨Ïä§ÌÑ∞ Í∞ÑÏùò Í±∞Î¶¨ (1 - MCS Ïú†ÏÇ¨ÎèÑ Í∏∞Î∞ò)
# ÏÑ∏Î°úÏÑ†: ÌÅ¥Îü¨Ïä§ÌÑ∞ Î≥ëÌï©
# ÏàòÌèâÏÑ†: ÌÅ¥Îü¨Ïä§ÌÑ∞ Í∞Ñ Í≤∞Ìï© Í±∞Î¶¨ (ÎÜíÏùÑÏàòÎ°ù Îçú Ïú†ÏÇ¨)
# Îπ®Í∞Ñ Ï†êÏÑ†(threshold=0.4): Ìï¥Îãπ Í±∞Î¶¨ÏóêÏÑú ÌÅ¥Îü¨Ïä§ÌÑ∞Î•º ÏûêÎ•¥Îäî Í∏∞Ï§ÄÏÑ†
# threshold=0.4	Ïú†ÏÇ¨ÎèÑ Í∏∞Ï§Ä :  0.6 Ïù¥ÏÉÅ(1 - 0.4)Ïù∏ Î∂ÑÏûêÎì§ÏùÑ Í∞ôÏùÄ ÌÅ¥Îü¨Ïä§ÌÑ∞Î°ú Î¨∂Ïùå
# ÏàòÌèâÏÑ† ÏïÑÎûòÏùò ÏÑ∏Î°ú Î¨∂ÏùåÎì§ :	Í∞Å ÌÅ¥Îü¨Ïä§ÌÑ∞ Í∑∏Î£π
# Ïö∞Ï∏°(ÎÖπÏÉâ)Ïùò Ï°∞Î∞ÄÌïú Í∑∏Î£π : ÏÑúÎ°ú Îß§Ïö∞ Ïú†ÏÇ¨Ìïú Íµ¨Ï°∞Î•º Í∞ÄÏßÑ Î∂ÑÏûêÍµ∞
# Ï¢åÏ∏°(ÌååÎûÄÏÉâ)Ïùò ÎÑìÏùÄ Í∞ÑÍ≤© : Íµ¨Ï°∞Ï†ÅÏúºÎ°ú Ïù¥ÏßàÏ†ÅÏù∏ Î∂ÑÏûêÎì§, Ïú†ÏÇ¨ÏÑ±Ïù¥ ÎÇÆÏùå



# PCA on MCS Similarity ÏãúÍ∞ÅÌôî Ìï¥ÏÑù
# PCA1, PCA2: MCS Ïú†ÏÇ¨ÎèÑ Í∏∞Î∞ò Í±∞Î¶¨ ÌñâÎ†¨ÏùÑ Ï†ÄÏ∞®Ïõê(2D)ÏúºÎ°ú Ï∂ïÏÜåÌïú Ï∂ï
# Ï†ê ÌïòÎÇò = ÌïòÎÇòÏùò Î∂ÑÏûê
# ÏÉâÏÉÅ = Ï†ÄÌï¥Ïú®(Inhibition) ÏàòÏ§Ä

# ÎπÑÏä∑Ìïú ÏúÑÏπòÏóê ÏûàÎäî Ï†êÎì§ = Íµ¨Ï°∞Ï†ÅÏúºÎ°ú Ïú†ÏÇ¨Ìïú Î∂ÑÏûêÎì§
# ÏÉâÏÉÅ Î∂ÑÌè¨Í∞Ä ÏùºÏ†ïÌïòÏßÄ ÏïäÏùå : Íµ¨Ï°∞ÎßåÏúºÎ°ú ÏôÑÏ†ÑÌûà ÏÑ§Î™ÖÎêòÏßÄ ÏïäÎäî Ï†ÄÌï¥Ïú® Îã§ÏñëÏÑ±Ïù¥ Ï°¥Ïû¨Ìï®]
# Í≥†Ï†ÄÌï¥Ïú® Î∂ÑÌè¨ ÌôïÏù∏ : Î¨¥ÏûëÏúÑÏ†ÅÏúºÎ°ú ÏÉâÏÉÅÏù¥ Î∂ÑÌè¨, MCS Íµ¨Ï°∞ÎßåÏúºÎ°ú Ï†ÄÌï¥Ïú®ÏùÑ ÏòàÏ∏°ÌïòÎäî Îç∞Îäî ÌïúÍ≥ÑÍ∞Ä ÏûàÏùå
```

```python
# ======================
# ÏÖÄ 3: Î¨ºÎ¶¨¬∑ÌôîÌïô ÌîºÏ≤ò Í≥ÑÏÇ∞ Ìï®Ïàò Ï†ïÏùò
# ======================
def featurize_molecule(mol):
    """Îã®Ïùº Î∂ÑÏûêÏóê ÎåÄÌï¥ Î∂ÑÏûêÎüâ, Í≥†Î¶¨ Ïàò, ÏõêÏûê Í∞úÏàò, Í∏∞Îä•ÏÑ± Í∑∏Î£π Í∞úÏàòÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§."""
    feats = {
        'MolWt': Descriptors.MolWt(mol),
        'NumCharges': sum(a.GetFormalCharge()!=0 for a in mol.GetAtoms()),
        'NumRings': rdMolDescriptors.CalcNumRings(mol),
        'NumAromaticRings': rdMolDescriptors.CalcNumAromaticRings(mol)
    }
    atom_types = ['C','H','N','O','Cl','S','F','Br','I','B','Si']
    for atom in atom_types:
        feats[atom] = sum(a.GetSymbol()==atom for a in mol.GetAtoms())

    fg_smarts = {
        'Amine_primary':'[NX3;H2]', 'Amine_secondary':'[NX3;H1]', 'Amine_tertiary':'[NX3;H0]',
        'Alcohol':'[OX2H]', 'Aldehyde':'[CX3H1](=O)[#6]', 'Ketone':'[#6][CX3](=O)[#6]',
        'Carboxylic_acid':'[CX3](=O)[OX2H1]', 'Ester':'[CX3](=O)[OX2][#6]', 'Nitrile':'[CX2]#N'
    }
    for name,patt in fg_smarts.items():
        feats[name] = len(mol.GetSubstructMatches(Chem.MolFromSmarts(patt)))
    return feats

# ÌîºÏ≤ò Í≥ÑÏÇ∞ Î∞è Î≥ëÌï©
test_feats = [featurize_molecule(m) for m in df['mol']]
feat_df = pd.DataFrame(test_feats)
df_features = pd.concat([df[['Canonical_Smiles','Inhibition']], feat_df], axis=1)
print(df_features.head())
print(f"ÌîºÏ≤ò Îß§Ìä∏Î¶≠Ïä§ ÌÅ¨Í∏∞: {df_features.shape}\n")
```

```python
# ======================
# ÏÖÄ 5: Morgan ÏßÄÎ¨∏ ÏÉùÏÑ±
# ======================
def get_morgan_fp(mol, radius=2, nBits=2048):
    return AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)

fps_morgan=[get_morgan_fp(m) for m in df['mol']]
print(f"Morgan fingerprints: {len(fps_morgan)}Í∞ú, Í∏∏Ïù¥={fps_morgan[0].GetNumBits()}")
fps_array=np.array([list(fp) for fp in fps_morgan])

# ÏÉùÏÑ±Îêú Ïàò: 1681Í∞ú ‚Üí Ï†ÑÏ≤¥ Î∂ÑÏûê ÏàòÏôÄ ÏùºÏπò (train set Í∏∞Ï§Ä)
# ÏßÄÎ¨∏ Í∏∏Ïù¥: 2048 ‚Üí ÏùºÎ∞òÏ†ÅÏù∏ ÏÑ§Ï†ï (ÎåÄÎ∂ÄÎ∂Ñ QSAR ÎåÄÌöåÏóêÏÑú 1024 ÎòêÎäî 2048 ÏÇ¨Ïö©
```

```python
# ======================
# ÏÖÄ 6: Atom-Pair ÏßÄÎ¨∏ Ï§ëÎ≥µ ÌÉêÏßÄ
# ======================
from rdkit.Chem import rdMolDescriptors
from collections import Counter

fps_ap=[rdMolDescriptors.GetHashedAtomPairFingerprintAsBitVect(m,nBits=2048) for m in df['mol']]
bitstrs=[fp.ToBitString() for fp in fps_ap]
counter=Counter(bitstrs)
dup_idxs=[i for i,bs in enumerate(bitstrs) if counter[bs]>1]
print(f"Ï§ëÎ≥µ AP ÏßÄÎ¨∏ ÌÅ¥Îü¨Ïä§ÌÑ∞: {len(set(bitstrs[i] for i in dup_idxs))}Í∞ú, Ï¥ù Ï§ëÎ≥µ={len(dup_idxs)}")
labels=[counter[bs] for bs in bitstrs]
print(pd.Series(labels).value_counts()[lambda x: x>2])

# Î™®Îì† ÏßÄÎ¨∏Ïùò Îì±Ïû• ÌöüÏàò	1ÌöåÏî© (Ï¶â, Ï†ÑÎ∂Ä Ïú†ÏùºÌï®)
```

```python
# ======================
# ÏÖÄ 7: MCS ÏÉòÌîåÎßÅ Î∞è ÏãúÍ∞ÅÌôî
# ======================
random.seed(42)
pairs=[tuple(random.sample(range(len(df)),2)) for _ in range(5)]

mols,highlights,legends=[],[],[]
for i,j in pairs:
    m1,m2=df.loc[i,'mol'],df.loc[j,'mol']
    res=rdFMCS.FindMCS([m1,m2])
    mcs_mol=Chem.MolFromSmarts(res.smartsString)
    match1=m1.GetSubstructMatch(mcs_mol)
    match2=m2.GetSubstructMatch(mcs_mol)
    mols+=[m1,m2]
    highlights+=[list(match1),list(match2)]
    legends+=[f"Mol{i}",f"Mol{j}"]

from rdkit.Chem.Draw import MolsToGridImage
img=MolsToGridImage(mols,highlightAtomLists=highlights,molsPerRow=2,subImgSize=(250,250),legends=legends)
display(img)
```

```python
# ‚úÖ 0. Import
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors
from rdkit.DataStructs import TanimotoSimilarity
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import AgglomerativeClustering
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
import warnings
warnings.filterwarnings("ignore")

# ‚úÖ 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Mol ÏÉùÏÑ±
train = pd.read_csv("/content/data_1/train.csv")
test = pd.read_csv("/content/data_1/test.csv")
sub = pd.read_csv("/content/data_1/sample_submission.csv")

train["mol"] = train["Canonical_Smiles"].map(Chem.MolFromSmiles)
test["mol"] = test["Canonical_Smiles"].map(Chem.MolFromSmiles)

train = train.dropna(subset=["mol"]).reset_index(drop=True)
test = test.dropna(subset=["mol"]).reset_index(drop=True)

# ‚úÖ 2. Î¨ºÎ¶¨ÌôîÌïô ÌîºÏ≤ò Ï∂îÏ∂ú
physchem_cols = ['MolWt','LogP','HDonors','HAcceptors','TPSA','RotBonds','Rings','CSP3','HeavyAtoms','MR']
def extract_physchem_features(mol):
    return [
        Descriptors.MolWt(mol),
        Descriptors.MolLogP(mol),
        Descriptors.NumHDonors(mol),
        Descriptors.NumHAcceptors(mol),
        Descriptors.TPSA(mol),
        Descriptors.NumRotatableBonds(mol),
        Descriptors.RingCount(mol),
        Descriptors.FractionCSP3(mol),
        Descriptors.HeavyAtomCount(mol),
        Descriptors.MolMR(mol)
    ]
train_phys = pd.DataFrame([extract_physchem_features(m) for m in train["mol"]], columns=physchem_cols)
test_phys = pd.DataFrame([extract_physchem_features(m) for m in test["mol"]], columns=physchem_cols)

# ‚úÖ 3. Morgan Fingerprint
def get_morgan_fp(mol, radius=2, nBits=512):
    return AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits)
train_fp = [get_morgan_fp(m) for m in train["mol"]]

# ‚úÖ 4. Ïú†ÏÇ¨ÎèÑ ÌñâÎ†¨ Î∞è Íµ∞Ïßë
n = len(train_fp)
sim_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(i + 1, n):
        sim = TanimotoSimilarity(train_fp[i], train_fp[j])
        dist = 1 - sim
        sim_matrix[i, j] = sim_matrix[j, i] = dist

agg = AgglomerativeClustering(n_clusters=10, metric='precomputed', linkage='average')
train["Cluster"] = agg.fit_predict(sim_matrix)

# ‚úÖ 5. ÌÅ¥Îü¨Ïä§ÌÑ∞ OHE + ÌîºÏ≤ò Í≤∞Ìï©
ohe = OneHotEncoder(sparse_output=False)
cluster_ohe = ohe.fit_transform(train[["Cluster"]])
cluster_ohe_test = np.zeros((test.shape[0], cluster_ohe.shape[1]))  # ÌÖåÏä§Ìä∏ÏÖãÏóêÎäî ÌÅ¥Îü¨Ïä§ÌÑ∞ ÏóÜÏùå

X_train = pd.concat([train_phys, pd.DataFrame(cluster_ohe)], axis=1)
X_test = pd.concat([test_phys, pd.DataFrame(cluster_ohe_test)], axis=1)
y_train = train["Inhibition"]

# ‚úÖ 6. Ïª¨ÎüºÎ™Ö ÌÜµÏùº Î∞è Ïä§ÏºÄÏùºÎßÅ
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ‚úÖ 7. ÏïôÏÉÅÎ∏î Î™®Îç∏ ÌïôÏäµ Î∞è ÏòàÏ∏°
models = [
    ("rf", RandomForestRegressor(n_estimators=200, random_state=42)),
    ("gb", GradientBoostingRegressor(n_estimators=200, learning_rate=0.05, random_state=42)),
    ("ridge", Ridge(alpha=1.0))
]

preds = np.zeros((X_test.shape[0], len(models)))
for i, (name, model) in enumerate(models):
    model.fit(X_train, y_train)
    preds[:, i] = model.predict(X_test)

# ‚úÖ 8. ÌèâÍ∑† ÏïôÏÉÅÎ∏î
final_preds = preds.mean(axis=1)

# ‚úÖ 9. Ï†úÏ∂ú
sub["Inhibition"] = final_preds
sub.to_csv("submission_clustered_ensemble.csv", index=False)
print("‚úÖ Ï†úÏ∂ú ÌååÏùº Ï†ÄÏû• ÏôÑÎ£å: submission_clustered_ensemble.csv")
```

```python
# üì¶ Î™®Îìà Î∂àÎü¨Ïò§Í∏∞
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, AllChem, Crippen
from rdkit.DataStructs import TanimotoSimilarity
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error
from joblib import Parallel, delayed
from tqdm import tqdm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# ‚úÖ Í≤ΩÎ°ú ÏÑ§Ï†ï
BASE_PATH = "/content/data_1/"
train = pd.read_csv(BASE_PATH + "train.csv")
test = pd.read_csv(BASE_PATH + "test.csv")
submission = pd.read_csv(BASE_PATH + "sample_submission.csv")

# ‚úÖ Mol Î≥ÄÌôò Ìï®Ïàò (None Î∞©ÏßÄ)
def safe_mol(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        return mol if mol is not None else Chem.MolFromSmiles('')
    except:
        return Chem.MolFromSmiles('')

train['Mol'] = train['Canonical_Smiles'].apply(safe_mol)
test['Mol'] = test['Canonical_Smiles'].apply(safe_mol)

# ‚úÖ Morgan Fingerprint Î≥ëÎ†¨ ÏÉùÏÑ±
def get_fp(mol):
    return AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)

fps_train = Parallel(n_jobs=-1)(delayed(get_fp)(mol) for mol in train['Mol'])
fps_test = Parallel(n_jobs=-1)(delayed(get_fp)(mol) for mol in test['Mol'])

# ‚úÖ Tanimoto Ïú†ÏÇ¨ÎèÑ ÌñâÎ†¨
def compute_similarity_matrix(fps):
    return np.array([
        [TanimotoSimilarity(fps[i], fps[j]) for j in range(len(fps))]
        for i in tqdm(range(len(fps)), desc="Fingerprint Similarity")
    ])

sim_fp_train = compute_similarity_matrix(fps_train)

# ‚úÖ PCA Î≥ÄÌôò
pca_fp = PCA(n_components=2, random_state=42)
fp_pca_train = pca_fp.fit_transform(sim_fp_train)
train['FP_PCA1'], train['FP_PCA2'] = fp_pca_train[:, 0], fp_pca_train[:, 1]

# ‚úÖ Test FP ‚Üí Train Í∏∞Î∞ò Ïú†ÏÇ¨ÎèÑ PCA
sim_fp_test = np.array([
    [TanimotoSimilarity(fp_t, fp_tr) for fp_tr in fps_train]
    for fp_t in tqdm(fps_test, desc="FP Similarity for Test")
])
fp_pca_test = pca_fp.transform(sim_fp_test)
test['FP_PCA1'], test['FP_PCA2'] = fp_pca_test[:, 0], fp_pca_test[:, 1]

# ‚úÖ RDKit Descriptor Ï∂îÏ∂ú
def extract_descriptors(df):
    df['MolWt'] = df['Mol'].apply(Descriptors.MolWt)
    df['TPSA'] = df['Mol'].apply(Descriptors.TPSA)
    df['LogP'] = df['Mol'].apply(Crippen.MolLogP)
    df['NumHDonors'] = df['Mol'].apply(Descriptors.NumHDonors)
    df['NumHAcceptors'] = df['Mol'].apply(Descriptors.NumHAcceptors)
    df['NumRotatableBonds'] = df['Mol'].apply(Descriptors.NumRotatableBonds)
    return df

train = extract_descriptors(train)
test = extract_descriptors(test)

# ‚úÖ ÏµúÏ¢Ö Feature ÏÑ†ÌÉù
features = ['MolWt', 'TPSA', 'LogP', 'NumHDonors', 'NumHAcceptors', 'NumRotatableBonds', 'FP_PCA1', 'FP_PCA2']
X = train[features]
y = train['Inhibition']
X_test = test[features]

# ‚úÖ Î™®Îç∏ ÌïôÏäµ Î∞è ÏòàÏ∏°
kf = KFold(n_splits=5, shuffle=True, random_state=42)
val_preds, val_trues = [], []
test_preds = np.zeros(len(test))

for train_idx, val_idx in kf.split(X):
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = RandomForestRegressor(
        n_estimators=500,
        max_depth=15,
        min_samples_leaf=2,
        min_samples_split=5,
        random_state=42,
        n_jobs=-1
    )
    model.fit(X_tr, y_tr)

    val_pred = model.predict(X_val)
    test_pred = model.predict(X_test)

    val_preds.extend(val_pred)
    val_trues.extend(y_val)
    test_preds += test_pred / kf.n_splits

# ‚úÖ ÏÑ±Îä• ÌèâÍ∞Ä
mae = mean_absolute_error(val_trues, val_preds)
rmse = np.sqrt(mean_squared_error(val_trues, val_preds))
print(f"‚úÖ MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# ‚úÖ Feature Importance ÏãúÍ∞ÅÌôî
model.fit(X, y)  # Ï†ÑÏ≤¥ ÌïôÏäµ ÌõÑ Ï§ëÏöîÎèÑ
importances = model.feature_importances_
feat_df = pd.DataFrame({'feature': features, 'importance': importances})
feat_df.sort_values(by='importance', ascending=False).plot.bar(x='feature', y='importance', title="Feature Importance", figsize=(8,5))
plt.tight_layout()
plt.show()

# ‚úÖ Ï†úÏ∂ú ÌååÏùº Ï†ÄÏû•
submission['Inhibition'] = test_preds
submission.to_csv(BASE_PATH + "submission_rf_final.csv", index=False)
print("üìÅ Ï†úÏ∂ú ÏôÑÎ£å: submission_rf_final.csv")
```

```python
# üì¶ ÌïÑÏàò Î™®Îìà Î°úÎìú
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, MACCSkeys, AllChem
from rdkit.DataStructs import ConvertToNumpyArray, TanimotoSimilarity
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostRegressor, Pool
import optuna
import shap
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# ‚úÖ Í≤ΩÎ°ú ÏÑ§Ï†ï
train = pd.read_csv("/content/data_1/train.csv")
test = pd.read_csv("/content/data_1/test.csv")
submission = pd.read_csv("/content/data_1/sample_submission.csv")

# ‚úÖ Î∂ÑÏûê Íµ¨Ï°∞ ÏÉùÏÑ±
train['Mol'] = train['Canonical_Smiles'].apply(Chem.MolFromSmiles)
test['Mol'] = test['Canonical_Smiles'].apply(Chem.MolFromSmiles)

# ‚úÖ Î∂ÑÏûê ÌäπÏÑ± Ï∂îÏ∂ú Ìï®Ïàò
def extract_features(df):
    df['MolWt'] = df['Mol'].apply(Descriptors.MolWt)
    df['TPSA'] = df['Mol'].apply(Descriptors.TPSA)
    df['LogP'] = df['Mol'].apply(Descriptors.MolLogP)
    df['NumHDonors'] = df['Mol'].apply(Descriptors.NumHDonors)
    df['NumHAcceptors'] = df['Mol'].apply(Descriptors.NumHAcceptors)
    df['NumRotatableBonds'] = df['Mol'].apply(Descriptors.NumRotatableBonds)

    maccs = []
    morgan = []
    for mol in df['Mol']:
        fp = MACCSkeys.GenMACCSKeys(mol)
        fp_arr = np.zeros((fp.GetNumBits(),))
        ConvertToNumpyArray(fp, fp_arr)
        maccs.append(fp_arr)

        morgan_fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=256)
        morgan_arr = np.zeros((256,))
        ConvertToNumpyArray(morgan_fp, morgan_arr)
        morgan.append(morgan_arr)

    maccs = np.array(maccs)
    morgan = np.array(morgan)
    maccs_df = pd.DataFrame(maccs, columns=[f'MACCS_{i}' for i in range(maccs.shape[1])])
    morgan_df = pd.DataFrame(morgan, columns=[f'Morgan_{i}' for i in range(morgan.shape[1])])

    df = pd.concat([df.reset_index(drop=True), maccs_df, morgan_df], axis=1)
    return df

train = extract_features(train)
test = extract_features(test)

# ‚úÖ ÌïôÏäµ Ï§ÄÎπÑ
target = train["Inhibition"]
features = [col for col in train.columns if col not in ['ID', 'Canonical_Smiles', 'Mol', 'Inhibition']]

X = train[features]
X_test = test[features]
y = target

# ‚úÖ Optuna ÌäúÎãù Ìï®Ïàò Ï†ïÏùò
def objective(trial):
    params = {
        "iterations": trial.suggest_int("iterations", 300, 800),
        "depth": trial.suggest_int("depth", 4, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.2),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1e-3, 10.0, log=True),
        "random_strength": trial.suggest_float("random_strength", 0.1, 1.0),
        "bagging_temperature": trial.suggest_float("bagging_temperature", 0.0, 1.0),
        "border_count": trial.suggest_int("border_count", 32, 254),
        "task_type": "GPU",
        "verbose": 0,
        "random_seed": 42
    }

    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    maes = []
    for train_idx, valid_idx in cv.split(X):
        X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
        y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

        model = CatBoostRegressor(**params)
        model.fit(X_train, y_train, eval_set=(X_valid, y_valid), early_stopping_rounds=50, verbose=0)
        preds = model.predict(X_valid)
        mae = mean_absolute_error(y_valid, preds)
        maes.append(mae)

    return np.mean(maes)

# ‚úÖ Optuna ÏàòÌñâ
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=30)
best_params = study.best_params
best_params["task_type"] = "GPU"
best_params["random_seed"] = 42
best_params["verbose"] = 0

# ‚úÖ ÏµúÏ¢Ö Î™®Îç∏ ÌïôÏäµ & ÏòàÏ∏°
kf = KFold(n_splits=5, shuffle=True, random_state=42)
oof_preds = np.zeros(len(train))
test_preds = np.zeros(len(test))

for fold, (tr_idx, val_idx) in enumerate(kf.split(X)):
    X_tr, y_tr = X.iloc[tr_idx], y.iloc[tr_idx]
    X_val, y_val = X.iloc[val_idx], y.iloc[val_idx]

    model = CatBoostRegressor(**best_params)
    model.fit(X_tr, y_tr, eval_set=(X_val, y_val), early_stopping_rounds=50)

    oof_preds[val_idx] = model.predict(X_val)
    test_preds += model.predict(X_test) / kf.n_splits

# ‚úÖ ÏÑ±Îä• Ï∂úÎ†•
mae = mean_absolute_error(y, oof_preds)
print(f"‚úÖ OOF MAE: {mae:.4f}")

# ‚úÖ SHAP Ìï¥ÏÑù
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
shap.summary_plot(shap_values, X, show=False)
plt.tight_layout()
plt.savefig("shap_summary.png")
print("üìä SHAP Summary Ï†ÄÏû• ÏôÑÎ£å: shap_summary.png")

# ‚úÖ Ï†úÏ∂ú ÌååÏùº ÏÉùÏÑ±
submission['Inhibition'] = test_preds
submission.to_csv("submission_catboost_all_features.csv", index=False)
print("üìÅ ÏµúÏ¢Ö Ï†úÏ∂ú ÌååÏùº Ï†ÄÏû• ÏôÑÎ£å: submission_catboost_all_features.csv")
```

```python
# ‚úÖ ÎùºÏù¥Î∏åÎü¨Î¶¨
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error

# ‚úÖ Í≤ΩÎ°ú ÏÑ§Ï†ï
train = pd.read_csv("/content/data_1/train.csv")
test = pd.read_csv("/content/data_1/test.csv")
submission = pd.read_csv("/content/data_1/sample_submission.csv")

# ‚úÖ ÎîîÎ∞îÏù¥Ïä§ ÏÑ§Ï†ï
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("‚úÖ Device:", device)

# ‚úÖ ChemBERTa Î°úÎî©
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
chemberta = AutoModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
chemberta.to(device)
chemberta.eval()

# ‚úÖ ChemBERTa ÏûÑÎ≤†Îî© Ï∂îÏ∂ú
def get_chemberta_embedding(smiles):
    with torch.no_grad():
        tokens = tokenizer(smiles, return_tensors='pt', padding='max_length', truncation=True, max_length=128).to(device)
        outputs = chemberta(**tokens)
        return outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy()

print("‚úÖ ChemBERTa ÏûÑÎ≤†Îî© Ï§ë...")
train_embed = np.vstack([get_chemberta_embedding(s) for s in train["Canonical_Smiles"]])
test_embed = np.vstack([get_chemberta_embedding(s) for s in test["Canonical_Smiles"]])
print("‚úÖ ÏûÑÎ≤†Îî© ÏôÑÎ£å!")

# ‚úÖ Dataset ÌÅ¥ÎûòÏä§
class SmilesDataset(Dataset):
    def __init__(self, X, y=None):
        self.X = X
        self.y = y
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        if self.y is not None:
            return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)
        else:
            return torch.tensor(self.X[idx], dtype=torch.float32)

# ‚úÖ MLP Î™®Îç∏ Ï†ïÏùò
class MLP(nn.Module):
    def __init__(self, input_dim=768):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 1)
        )
    def forward(self, x):
        return self.net(x)

# ‚úÖ BERT Í∏∞Î∞ò MLP Î™®Îç∏ ÌïôÏäµ
def train_bert_model(X, y, X_test, n_splits=5, epochs=10, batch_size=32):
    preds = np.zeros(len(X_test))
    oof = np.zeros(len(X))
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    for fold, (tr_idx, va_idx) in enumerate(kf.split(X)):
        print(f"üîÅ Fold {fold+1}")
        model = MLP().to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
        loss_fn = nn.MSELoss()

        train_loader = DataLoader(SmilesDataset(X[tr_idx], y[tr_idx]), batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(SmilesDataset(X[va_idx], y[va_idx]), batch_size=batch_size)
        test_loader = DataLoader(SmilesDataset(X_test), batch_size=batch_size)

        for epoch in range(epochs):
            model.train()
            for xb, yb in train_loader:
                xb, yb = xb.to(device), yb.to(device)
                optimizer.zero_grad()
                pred = model(xb).squeeze()
                loss = loss_fn(pred, yb)
                loss.backward()
                optimizer.step()

        # ‚úÖ Validation
        model.eval()
        val_preds = []
        with torch.no_grad():
            for xb, _ in val_loader:
                xb = xb.to(device)
                val_preds.extend(model(xb).squeeze().cpu().numpy())
        oof[va_idx] = val_preds

        # ‚úÖ Test
        test_preds = []
        with torch.no_grad():
            for xb in test_loader:
                xb = xb.to(device)
                test_preds.extend(model(xb).squeeze().cpu().numpy())
        preds += np.array(test_preds) / n_splits

    return oof, preds

# ‚úÖ MLP ÌïôÏäµ
y = train["Inhibition"].values
oof_bert, pred_bert = train_bert_model(train_embed, y, test_embed)

# ‚úÖ CatBoost ÌïôÏäµ (ÏûÑÎ≤†Îî© ÏÇ¨Ïö©)
cat_model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    task_type='GPU',
    devices='0',
    verbose=0
)
cat_model.fit(train_embed, y)
oof_cat = cat_model.predict(train_embed)
pred_cat = cat_model.predict(test_embed)

# ‚úÖ ÏïôÏÉÅÎ∏î (ÌèâÍ∑†)
final_oof = (oof_bert + oof_cat) / 2
final_pred = (pred_bert + pred_cat) / 2

# ‚úÖ ÌèâÍ∞Ä Î∞è Ï†úÏ∂ú
mae = mean_absolute_error(y, final_oof)
print(f"‚úÖ ÏµúÏ¢Ö ÏïôÏÉÅÎ∏î OOF MAE: {mae:.4f}")

submission["Inhibition"] = final_pred
submission.to_csv("/content/data_1/submission_ensemble_chemberta_catboost.csv", index=False)
print("‚úÖ Ï†úÏ∂ú ÌååÏùº Ï†ÄÏû• ÏôÑÎ£å!")
```

```python
# ‚úÖ Î™®Îìà ÏûÑÌè¨Ìä∏
import pandas as pd, numpy as np
import torch, shap, optuna
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
from rdkit import Chem
from rdkit.Chem import Descriptors, AllChem, MACCSkeys, Lipinski
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from transformers import RobertaTokenizer, RobertaModel
import warnings
warnings.filterwarnings("ignore")

# ‚úÖ Îç∞Ïù¥ÌÑ∞ Î°úÎìú
train = pd.read_csv("/content/data_1/train.csv")
test = pd.read_csv("/content/data_1/test.csv")
submission = pd.read_csv("/content/data_1/sample_submission.csv")

# ‚úÖ Mol ÏÉùÏÑ±
train["Mol"] = train["Canonical_Smiles"].apply(Chem.MolFromSmiles)
test["Mol"] = test["Canonical_Smiles"].apply(Chem.MolFromSmiles)

# ‚úÖ ÌîºÏ≤ò Ï∂îÏ∂ú Ìï®Ïàò (Desc + MACCS + Morgan)
def get_features(df):
    rows = []
    for mol in df["Mol"]:
        if mol is None:
            rows.append([np.nan] * (5 + 166 + 2048))
            continue
        desc = [
            Descriptors.MolWt(mol),
            Lipinski.RingCount(mol),
            Lipinski.NumAromaticRings(mol),
            Descriptors.NumHAcceptors(mol),
            Descriptors.NumHDonors(mol)
        ]
        maccs = list(MACCSkeys.GenMACCSKeys(mol))
        morgan = list(AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048))
        rows.append(desc + maccs + morgan)
    return pd.DataFrame(rows)

X_train_fp = get_features(train).fillna(0)
X_test_fp = get_features(test).fillna(0)
X_train_fp.columns = [f'feat_{i}' for i in range(X_train_fp.shape[1])]
X_test_fp.columns = X_train_fp.columns

# ‚úÖ ChemBERTa ÏûÑÎ≤†Îî©
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = RobertaTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
model = RobertaModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1").to(device)
model.eval()

def get_chemberta_embedding(smiles_list):
    embeddings = []
    with torch.no_grad():
        for smi in smiles_list:
            inputs = tokenizer(smi, return_tensors="pt", truncation=True, padding="max_length", max_length=128).to(device)
            outputs = model(**inputs)
            emb = outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy()
            embeddings.append(emb)
    return np.array(embeddings)

X_train_bert = get_chemberta_embedding(train["Canonical_Smiles"].tolist())
X_test_bert = get_chemberta_embedding(test["Canonical_Smiles"].tolist())

# ‚úÖ ÏµúÏ¢Ö feature Í≤∞Ìï©
X_train_all = np.concatenate([X_train_fp.values, X_train_bert], axis=1)
X_test_all = np.concatenate([X_test_fp.values, X_test_bert], axis=1)
y = train["Inhibition"]

# ‚úÖ Optuna ÌäúÎãùÏö© Ìï®Ïàò (GPU Ìè¨Ìï®)
def optuna_objective(model_type):
    def objective(trial):
        if model_type == "cat":
            params = {
                "depth": trial.suggest_int("depth", 4, 10),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
                "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1, 10),
                "task_type": "GPU",
                "devices": "0",
                "random_seed": 42,
                "verbose": 0
            }
            model = CatBoostRegressor(**params)
        elif model_type == "lgb":
            params = {
                "num_leaves": trial.suggest_int("num_leaves", 31, 128),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
                "n_estimators": 1000,
                "device": "gpu",  # ‚úÖ GPU
                "gpu_use_dp": False,
                "random_state": 42
            }
            model = LGBMRegressor(**params)
        else:  # xgb
            params = {
                "max_depth": trial.suggest_int("max_depth", 4, 10),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
                "n_estimators": 1000,
                "reg_lambda": trial.suggest_float("reg_lambda", 1, 5),
                "tree_method": "gpu_hist",         # ‚úÖ GPU
                "predictor": "gpu_predictor",      # ‚úÖ GPU
                "random_state": 42
            }
            model = XGBRegressor(**params)

        scores = []
        for train_idx, val_idx in KFold(n_splits=5, shuffle=True, random_state=42).split(X_train_all):
            X_tr, X_val = X_train_all[train_idx], X_train_all[val_idx]
            y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]
            model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], early_stopping_rounds=30, verbose=0)
            preds = model.predict(X_val)
            scores.append(mean_absolute_error(y_val, preds))
        return np.mean(scores)
    return objective

# ‚úÖ Optuna ÌäúÎãù ÏàòÌñâ
for name in ["cat", "lgb", "xgb"]:
    study = optuna.create_study(direction="minimize")
    study.optimize(optuna_objective(name), n_trials=25)
    globals()[f"{name}_params"] = study.best_params

# ‚úÖ Î™®Îç∏ ÌïôÏäµ (GPU ÏÑ§Ï†ï Ìè¨Ìï®)
cat_model = CatBoostRegressor(**cat_params, task_type='GPU', devices='0', random_seed=42).fit(X_train_all, y)
lgb_model = LGBMRegressor(**lgb_params, device='gpu', gpu_use_dp=False, random_state=42).fit(X_train_all, y)
xgb_model = XGBRegressor(**xgb_params, tree_method='gpu_hist', predictor='gpu_predictor', random_state=42).fit(X_train_all, y)

# ‚úÖ ÏòàÏ∏° Î∞è ÏïôÏÉÅÎ∏î
cat_pred = cat_model.predict(X_test_all)
lgb_pred = lgb_model.predict(X_test_all)
xgb_pred = xgb_model.predict(X_test_all)

final_pred = (cat_pred + lgb_pred + xgb_pred) / 3
submission["Inhibition"] = final_pred
submission.to_csv("submission_ensemble_cat_lgb_xgb_gpu.csv", index=False)
print("‚úÖ Ï†ÄÏû• ÏôÑÎ£å: submission_ensemble_cat_lgb_xgb_gpu.csv")

# ‚úÖ SHAP ÏãúÍ∞ÅÌôî (CatBoost Í∏∞Ï§Ä)
explainer = shap.Explainer(cat_model)
shap_values = explainer(X_train_all[:300])
shap.summary_plot(shap_values, show=False)
plt.savefig("shap_summary_gpu.png")
```
