```python
# ✅ 0. 기본 라이브러리
import os
import pandas as pd
import numpy as np
import zipfile
import warnings
from rdkit import Chem
from rdkit.Chem import AllChem
from pycaret.regression import *
from rdkit import rdBase
rdBase.DisableLog('rdApp.error')
warnings.filterwarnings("ignore")

# ✅ 1. Google Drive 마운트 및 압축 해제
from google.colab import drive
drive.mount('/content/drive')

zip_path = '/content/drive/MyDrive/data.zip'
extract_dir = '/content/data'

if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print("✅ 압축 해제 완료:", extract_dir)

# ✅ 2. 데이터 로드
train_df = pd.read_csv(f"{extract_dir}/train.csv")
test_df = pd.read_csv(f"{extract_dir}/test.csv")
submission_df = pd.read_csv(f"{extract_dir}/sample_submission.csv")

# ✅ 3. 컬럼 정리
train_df.columns = train_df.columns.str.strip()
test_df.columns = test_df.columns.str.strip()

# ✅ 4. Morgan Fingerprint 함수
def get_morgan_fingerprint(smiles_string, n_bits=2048, radius=2):
    try:
        mol = Chem.MolFromSmiles(smiles_string)
        if mol is not None:
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
            return np.array(fp)
        else:
            return np.zeros(n_bits, dtype=int)
    except:
        return np.zeros(n_bits, dtype=int)

# ✅ 5. Fingerprint 생성
print("🔍 Morgan Fingerprint 추출 중...")
train_fps = train_df['Canonical_Smiles'].apply(get_morgan_fingerprint)
test_fps = test_df['Canonical_Smiles'].apply(get_morgan_fingerprint)

fp_columns = [f'FP_{i}' for i in range(2048)]
train_fp_df = pd.DataFrame(np.vstack(train_fps.values), columns=fp_columns)
test_fp_df = pd.DataFrame(np.vstack(test_fps.values), columns=fp_columns)
pycaret_train_df = pd.concat([train_fp_df, train_df['Inhibition']], axis=1)

# ✅ 6. PyCaret 환경 설정
print("⚙️ PyCaret 환경 설정 중...")
setup(data=pycaret_train_df,
      target='Inhibition',
      session_id=42,
      fold=5,
      normalize=True,
      normalize_method='minmax',
      verbose=False)

# ✅ 7. 모델 비교 및 선택
print("📊 모델 비교 중...")
best_model = compare_models(sort='RMSE', n_select=1)
results = pull()
print("✅ 모델 비교 결과:")
print(results.head())

# ✅ 8. 최종 모델 학습 및 예측
final_model = finalize_model(best_model)
preds = predict_model(final_model, data=test_fp_df)
final_preds = np.clip(preds['prediction_label'].values, 0, 100)

# ✅ 9. 제출 파일 생성
submission_df['Inhibition'] = final_preds
submission_df.to_csv('/content/submission_pycaret.csv', index=False)
print("✅ 제출 파일 저장 완료: /content/submission_pycaret.csv")
```
